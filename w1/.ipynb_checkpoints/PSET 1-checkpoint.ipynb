**Problem 1** 

Plot a lunar orbit using 50 time steps, `sin`, and `cos`. (Do not try to model the effects of gravity).

```python
import matplotlib.pyplot as plt
import numpy as np

lunar_distance = 384e6 # m
num_steps = 50

def orbit():
    position = np.zeros([num_steps, 2])
    for step in range(num_steps):
        position[step][0] = np.cos((2 * np.pi)/num_steps * step)
        position[step][1] = np.sin((2 * np.pi)/num_steps * step)
    return position

# todo - get intution for difference between number of steps and length of array
fig = plt.figure()
ax1 = fig.add_subplot(111)
ax1.plot(orbit()[:, 0], orbit()[:, 1], color='lightblue')
fig.show()
```
